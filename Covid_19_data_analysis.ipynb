# Q 2: Remove all the records where confirmed cases is less than 10

```python
#df.Confirmed <10
#df[df.Confirmed < 10]
#df[~df.Confirmed <10]
#df = df[~df.Confirmed < 10]
```

```python
data = data[~data.Confirmed <10]
```

```python
data
```

# Q3: In which Region, maximum number of confirmed cases were recorded? 

```python
data.groupby('Region').Confirmed.sum().sort_values(ascending=False).head(10)
```

# Q4: In which region, minimum number of deaths cases were recorded?

```python
data.groupby('Region').Deaths.sum().sort_values(ascending=True)
```

# Q5: How many confirmed, deaths and recovered cases were reported from till 29 april 2020?

```python
data[((data.Region == 'India') & (data.Date == '4/29/2020'))].sum()
```

# Q6(A): Sort the entire data wrt No. of confirmed cases in ascending order.

```python
data.sort_values(by=['Confirmed'], ascending=True)
```

# Q6(B): Sort the entire data wrt No. of Recovered cases in descending order.

```python
data.sort_values(by=['Recovered'], ascending=False)
```
